###### Query 7 - Total patients by Age and Category

Inference - Male of age 50+ have more chances of getting diagnosed with Colorectal cancer

```python
qq = """
with tb1 as (

select GA1.age, count(*) M, 0 F from (select age,gender from colorectal_data
group by patient_id) as GA1
group by GA1.age
having GA1.gender = 'M'

union all

select GA2.age, 0 M, count(*) F from (select age,gender from colorectal_data
group by patient_id) as GA2
group by GA2.age
having GA2.gender = 'F'

),
tb2 as (
select age, M, F, case when age in ('25-29','30-34','35-39','40-44','45-50') then '25-50'
                        when age in ('50-64','55-59','60-54','70-74','65-69') then '50-75'
                        else '75+'
                        end as Cat 
            from tb1
            )
select Cat Category, SUM(M) as Male, SUM(F) as Female, SUM(M)+SUM(F) as Total  from tb2
group by Cat
"""


ps.sqldf(qq, locals())
```

###### Query 8 - Top treated brands in each State
